NUTRITION DATA VISUALIZATION


```python
plt.figure()
sns.histplot(nutrition['calories_per_100g'], kde=True)
plt.title("Calories Distribution")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='protein_g', y='calories_per_100g', data=nutrition)
plt.title("Protein vs Calories")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


numeric_data = nutrition.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Nutrition Correlation")
plt.show()
```

BIOMETRIC DATA VISUALIZATION

```python
plt.figure()
sns.histplot(biometric['age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure()
sns.barplot(x='difficulty', y='calorie_burn_30min', data=exercise)
plt.title("Difficulty vs Calories Burned")
plt.show()
```

ADVANCED VISUALS

```python
plt.figure()
sns.violinplot(y=exercise['calorie_burn_30min'])
plt.title("Calories Burned Distribution")
plt.show()
```
